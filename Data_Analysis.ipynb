2. car_prices_in_rupee 

```python
data['car_prices_in_rupee'] = data['car_prices_in_rupee'].str.replace(',','')
```

```python
x = data['car_prices_in_rupee'][0]
print(x)
x[:x.index(' ')]
```

```python
def rupee_change(x):
    p = x.split(' ')
    try:
        if (p[1] == 'Lakh'):
            return str(round(float(p[0]) * 100000,1))
        elif (p[1] == 'Crore'):
            return str(round(float(p[0]) * 10000000,1))
    except:
        return x
```

```python
data['car_prices_in_rupee'] = data['car_prices_in_rupee'].apply(rupee_change)
```

```python
data.head(10)
```

```python
data.info()
```

change data type of car_prices_in_rupee column

```python
data['car_prices_in_rupee'] = data['car_prices_in_rupee'].astype('float')
```

```python
data.info()
```
